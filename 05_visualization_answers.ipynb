###1. Start by loading the `subway_stations.csv` to a DataFrame, and then use `matplotlib` to plot a `scatter` plot of `stop_lon` against the `stop_lat`. Also:

1.   Customize the scatter plot and change the marker `color` to green, and the transparency `alpha=0.3`.
2. Add labels, a title, a grid, and a legend.

```python
import pandas as pd
import matplotlib.pyplot as plt

subway = pd.read_csv('/content/data/subway_stations.csv')
subway.head()
```

```python
fig, ax = plt.subplots()
ax.scatter(x=subway['stop_lat'], y=subway['stop_lon'], color='green', alpha=0.3) 
ax.set_title('Longitude vs Latitude')
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
ax.set_axisbelow(True)
ax.grid(alpha=0.3)
```

###2. Plot the same data from (1) using `seaborn` instead of `matplotlib`.

```python
import seaborn as sns

sns_fig, sns_ax = plt.subplots()
sns.scatterplot(data=subway, x='stop_lat', y='stop_lon', ax=sns_ax, color='green', alpha=0.3).set(xlabel='Latitude', ylabel='Longitutde')
sns_ax.set_title('Longitude vs Latitude')
```
