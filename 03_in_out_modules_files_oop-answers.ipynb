### 2. Using `w`, create a file called `fruits.txt` and write the string "tangerines" to it. Next, use `a` to append the string "dragon fruit" to the file on a new line. Finally, use `r` to read and print the entire file. 

```python
with open('fruits.txt', 'w') as fruit_file:
    fruit_file.write('tangerines')

with open('fruits.txt', 'a') as fruit_file:
    fruit_file.write('\ndragon fruit')

with open('fruits.txt', 'r') as fruit_file:
    print(fruit_file.read())
```

### 3. Create a class called `Book` that has the attributes `title`, `author`, `pages`, `price`. It also has the method `getprice()` which returns `price`. Then create two instances of this class:

- `b1`:
`title` -> 'Practical Programming...'
`author` -> 'Gries, Campbell, Montojo'
`pages` -> 383
`price` -> 50

- `b2`:
`title` -> 'Building a Career in Data Science'
`author` -> 'Robinson, Nolis'
`pages` -> 322
`price` -> 40

```python
class Book:
    def __init__(self, title, author, pages, price):
        self.title = title
        self.author = author
        self.pages = pages
        self.price = price
    def getprice(self):
        return self.price

b1 = Book('Practical Programming...', 'Gries, Campbell, Montojo', 383, 50)
b2 = Book('Building a Career in Data Science', 'Robinson, Nolis', 322, 40)

print(b1.getprice())
```
