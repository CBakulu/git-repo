### 1. Write a function named `different` that takes two arguments, `a` and `b`. If `a` and `b` are different values, the function should return `True`. Otherwise it should return `False`.

HINT: You can write this function with only one line of code -- the `return` statement -- in the body.

```python
def different(a, b):
    '''
    Tell if a and b are different values.
    >>> different(3, 3.0)
    False
    >>> different(3, '3')
    True
    >>> different('apple', 'potato')
    True
    '''
    return a != b
```

### 2. Write a conditional that prints different messages if a bank account balance is below `$`3,000, between `$`3,000 and `$`10,000, or over `$`10,000.

HINT: You can check if a value `x` is between two other values with a condition like `5 <= x <= 10`, but you can also solve this problem by strategically ordering conditions.

```python
#NOTE: The language here was a little vague -- you may have opted for the <=/>= comparison operators.

balance = 3000

if balance < 3000:
    print('Your balance is under $3,000')
elif balance <= 10000:
    print('Your balance is between $3,000 and $10,000')
else:
    print('Your balance is over $10,000.')
```

### 4. Given the list `people`, sort it by people's first name, last name and age. Store the sorted lists as `by_first_name`, `by_last_name`, and `by_age`, respectively.  
`people = [('Mark', 'Harrison', 56), ('Ken', 'Wolseley', 23), ('Emily', 'Robinson', 77)]`

HINT 1: We want our sorting function to return a new list, rather than modifying in place.

HINT 2: `sorted()` accepts a `key` argument. This argument is the name of a function to use when sorting list elements.

```python
#Here, we have to write functions to get the first, second, and third items in a list.
#Then, we can use those functions as the keys for sorted().

people = [('Mark', 'Harrison', 56), ('Ken', 'Wolseley', 23), ('Emily', 'Robinson', 77)]

def get_first_item(lst):
    return lst[0]

def get_second_item(lst):
    return lst[1]

def get_third_item(lst):
    return lst[2]

by_first_name = sorted(people, key=get_first_item)
by_last_name = sorted(people, key=get_second_item)
by_age = sorted(people, key=get_third_item)
```

### 5. Write a function called `dict_intersect` that takes two dictionaries, `d1` and `d2`, as arguments and returns a set that contains only the keys found in both of the original dictionaries.

HINT: Let's break this down into steps. We need to:
*   Get the keys in `d1`
*   Get the keys in  `d2`
*   Convert them both to sets
*   Find their intersection

HINT: Some useful functions and methods are `keys()`, `set()`, and `intersection()`.

```python
def dict_intersect(d1, d2):
    '''Return the set of keys found in both d1 and d2.
    >>> dict_intersect({'a': 'A', 'b': 'B', 'c': 'C'}, {'a': 'alpha', 'b': 'beta'})
    {'a', 'b'}
    >>> dict_intersect({'a': 1, 'b': 2}, {'c': 3, 'd': 2})
    set()
    '''
    d1_keys = set(d1.keys())
    d2_keys = set(d2.keys())
    return d1_keys.intersection(d2_keys)
```

### 6. Write a loop that iterates over the two lists below simultaneously. For each pair of values, print the first number divided by the second. When the program encounters a zero divisor, it should skip the pair without printing anything.  

HINT: We can exit a loop early with `break`.

HINT 2: Remember that we can bundle two lists pairwise with `zip()`.

```python
dividends = [100, 37.5, -12]
divisors = [8, 0, -3]

for x, y in zip(dividends, divisors):
    if y == 0:
        continue
    else:
        print(x/y)
```
